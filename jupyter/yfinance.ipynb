```python
# read tickers from company_tickers.json
import json
with open('company_tickers.json') as f:
    data = json.load(f)
    tickers = list(data.values())
    tickers = [t['ticker'] for t in tickers]

display("Total tickers: ", len(tickers))
display(pd.DataFrame(tickers))
```

```python
industries = {}

# get industry for first 10 tickers
for ticker in tickers[:10]:
    ticker_info = yf.Ticker(ticker).info
    industries[ticker] = ticker_info.get('industry', 'N/A')

display(pd.DataFrame(industries.items(), columns=['Ticker', 'Industry']))
```
